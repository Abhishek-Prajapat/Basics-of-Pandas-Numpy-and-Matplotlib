# Importing the dataset

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline 
# will plot even if plt.show() is not called

iris = pd.read_csv('iris.csv')
iris.head() # head gives the first 5 rows
```

```python
print(iris.shape)
```

```python
print(iris.columns)
```

```python
iris['species'].value_counts()
```

## 2D scatter plot 

```python
iris.plot(kind='scatter',x='sepal_length',y='sepal_width')
plt.grid()
```

```python
sns.set_style('whitegrid')
sns.FacetGrid(iris, hue='species', height=4, aspect=2)   \
    .map(plt.scatter, 'sepal_length', 'sepal_width')  \
    .add_legend();
# to get info '?sns.FacetGrid'
```

## Pair plots

```python
sns.set_style('whitegrid')
sns.pairplot(iris, hue='species', height=4, aspect=2)
```

## Histogram 

```python
iris_settosa = iris.loc[iris['species'] == 'setosa'];
iris_virginica = iris.loc[iris['species'] == 'virginica'];
iris_versicolor = iris.loc[iris['species'] == 'versicolor'];

plt.plot(iris_settosa['petal_length'], np.zeros_like(iris_settosa['petal_length']), color='blue')
plt.plot(iris_versicolor['petal_length'], np.zeros_like(iris_versicolor['petal_length']), color='red')
plt.plot(iris_virginica['petal_length'], np.zeros_like(iris_virginica['petal_length']), color= 'green')
```

```python
?sns.distplot
sns.FacetGrid(iris, hue='species', height=5) \
    .map(sns.distplot, 'petal_length') \
    .add_legend()
```

```python
counts, bin_edges = np.histogram(iris_settosa['petal_length'], bins=10,
                                density = True)
pdf = counts/(sum(counts))
print(pdf)
print(bin_edges)
```

```python
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
```

```python
counts1, bin_edges1 = np.histogram(iris_settosa['petal_length'], bins=10,
                                density = True)
pdf1 = counts1/(sum(counts1))

counts2, bin_edges2 = np.histogram(iris_virginica['petal_length'], bins=10,
                                density = True)
pdf2 = counts2/(sum(counts2))

counts3, bin_edges3 = np.histogram(iris_versicolor['petal_length'], bins=10,
                                density = True)
pdf3 = counts3/(sum(counts3))

cdf1 = np.cumsum(pdf1)
plt.plot(bin_edges1[1:],pdf1)
plt.plot(bin_edges1[1:],cdf1)

cdf2 = np.cumsum(pdf2)
plt.plot(bin_edges2[1:],pdf2)
plt.plot(bin_edges2[1:],cdf2)

cdf3 = np.cumsum(pdf3)
plt.plot(bin_edges3[1:],pdf3)
plt.plot(bin_edges3[1:],cdf3)
```
