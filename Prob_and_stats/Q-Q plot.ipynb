# Quantile-Quantile plot 

###### This is used to see wheather a measurement is gaussian or not.
###### If if roughly lies on the line then its guassian

```python
import numpy as np
import pylab
import scipy.stats as stats
```

```python
std_normal = np.random.normal(loc = 0, scale = 1, size=100)
# loc = mean, scale = std-dev
```

```python
for i in range(0,101):
    print(i, np.percentile(std_normal,i))
```

```python
measurements = np.random.normal(loc = 20, scale = 5, size=100)
```

```python
stats.probplot(measurements, dist='norm', plot=pylab)
pylab.show()
```
